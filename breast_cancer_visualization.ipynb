# Visualization

Import the necessary packages for analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the dataset that we'll be working with

```python
breast_df = pd.read_csv("breast_cancer.csv", index_col="Unnamed: 0")
breast_df.head(4)
```

Get an overview of the columns of the dataframe

```python
# a list of all the columns in the dataframe
breast_df.columns
```

Make copies of the dataframe for when the tumor was cancerous and noncancerous

```python
# make copies of the dataframe for cancerous and non-cancerous tumors
cancerous_df = breast_df.loc[breast_df["outcome"] == 1]
noncancerous_df = breast_df.loc[breast_df["outcome"] == 0]
```

To get an idea of the the percentage of cancerous and non-cancerous tumors in the dataset, Let's try visualizing the data on a pie chart

```python
fig, ax = plt.subplots()

# count the number of times tumors were cancerous and non-cancerous
cancerous_count, noncancerous_count = breast_df["outcome"].value_counts()

# plot a pie chart, showing the number of cancerous and noncancerous tumors in percent
ax.pie(x=[cancerous_count, noncancerous_count], labels=["Cancerous", "Noncancerous"], radius=1.5, autopct="%.2f %%")

# set the title for the chart
ax.set_title(label="Cancerous-Noncancerous in the dataset", pad=50, fontdict={"fontname":"Serif", "fontsize":20, "color":"#72a333"})

# show the plot
plt.show()
```

> - Looking at the pie chart, it's obvious that there are more cancerous tumors than noncancerous in the data set

```python
# set the dimensions of the plot figure
fig = plt.figure(figsize=(14, 6))

# create axes object and set the title of the plot
ax = plt.axes()

ax.set_title(label="Distribution of average radius of non-cancerous and cancerous Tumors", fontdict={"fontname":"Serif", "fontsize":20})

# cancerous kde plot
sns.kdeplot(data=cancerous_df["mean radius"], label="Cancerous", shade=True)
# noncancerous kde plot
sns.kdeplot(data=noncancerous_df["mean radius"], label="Non-cancerous", shade=True)

# show the legend
plt.legend()

# show the plot
plt.show()
```

From here it can be observed that the non-cancerous tumors have larger **average radius** values than the cancerous tumors

```python
# set the dimensions of the plot figure
fig = plt.figure(figsize=(14, 6))

# create axes object and set the title of the plot
ax = plt.axes()

ax.set_title(label="Distribution of average perimeter of non-cancerous and cancerous Tumors", fontdict={"fontname":"Serif", "fontsize":20})

# cancerous kde plot
sns.kdeplot(data=cancerous_df["mean perimeter"], label="Cancerous", shade=True)
# noncancerous kde plot
sns.kdeplot(data=noncancerous_df["mean perimeter"], label="Non-cancerous", shade=True)

# show the legend
plt.legend()

# show the plot
plt.show()
```

From the above plot it can also be deduced that non-cancerous tumors could have larger **average perimeter** values than non-cancerous tumors, which is actually expected judging from the inference drawn in the previous plot where non-cancerous tumors were observed to have larger **average radius** values.

```python
# set the dimensions of the plot figure
fig = plt.figure(figsize=(14, 6))

# create axes object and set the title of the plot
ax = plt.axes()

ax.set_title(label="Distribution of average concavity of non-cancerous and cancerous Tumors", fontdict={"fontname":"Serif", "fontsize":20})

# cancerous kde plot
sns.kdeplot(data=cancerous_df["mean concavity"], label="Cancerous", shade=True)
# noncancerous kde plot
sns.kdeplot(data=noncancerous_df["mean concavity"], label="Non-cancerous", shade=True)

# show the legend
plt.legend()

# show the plot
plt.show()
```

From the above plot it can be noticed that more cancerous tumors have a lower **average concavity** than non-cancerous ones.
